```python
##importing important dictionaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 1 : Reading and Understanding the data

```python
#reading a data
Bike=pd.read_csv('day.csv')
```

```python
#displaying head of the file
Bike.head(10)
```

```python
##looking at the data it can be seen  instant column represents row number so setting instant as index column
```

```python
Bike.set_index('instant',inplace=True)
```

```python
##getting brief information about bike
Bike.info()
```

```python
#from info it can be seen that there are no missing values in each variable as shape is (730,15) and number of non-null count is 730
```

```python
##getting brief discription of the dataset
Bike.describe()
```

```python
##checking unique values in all columns
Columns= Bike.columns
for i in Columns:
    UniValues=Bike[i].unique()
    print(i + " : " + str(UniValues))
```

```python
##from above unique values and data dictionary segmenting categorical variables and continuous variable for further use
CatVar=['season','yr','mnth','holiday','weekday','workingday','weathersit']
ContVar=['temp','atemp','hum','windspeed','casual','registered','cnt']
DateVar=['dteday']
```

```python
##checking correlation between variables
Correlation=Bike.corr()
Correlation
```

### Visualsing the data

```python
##plotting scatter plot for continuous variables
sns.pairplot(Bike[ContVar])
plt.show()
```

```python
###Plotting heatmap of all variables
plt.figure(figsize=(15,15))
sns.heatmap(Correlation,cmap="YlGnBu",annot=True)
plt.show()
```

```python
for i in CatVar:
    sns.boxplot(x=i, y='cnt', data=Bike)
    plt.title('Box Plot of Count Vs '+ i +' \n')
    plt.show()
```
